```python
airline=driver.find_elements("xpath",'//span[@class="i-b text ellipsis"]')
len(airline)
Airline=[]
for i in airline:
    Airline.append(i.text)
Airline

flight_no=driver.find_elements("xpath",'//p[@class="normal fs-11 font-lightestgrey no-wrap fl-no"]')
len(flight_no)

Flight_no=[]
for i in flight_no:
    Flight_no.append(i.text)
Flight_no

source=driver.find_elements('xpath','//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
len(source)

sources=[]
for i in source:
    sources.append(i.text)
sources

list1 = [sources[i] for i in range(len(sources)) if i % 2 != 0]
len(list1)
list1

list2=sources[::2]

len(list2)
list2

duration_tag=driver.find_elements("xpath",'//p[@class="fs-12 bold du mb-2"]')
len(duration_tag)

duration=[]
for i in duration_tag:
    duration.append(i.text)
duration

arr_t=driver.find_elements("xpath",'//div[@class="fs-15 bold time"]')
len(arr_t)

Arr_t=[]
for i in arr_t:
    Arr_t.append(i.text)
Arr_t

dep_t=driver.find_elements("xpath",'//p[@class="bold fs-15 mb-2 pr time"]')
len(dep_t)

Dep_t=[]
for i in dep_t:
    Dep_t.append(i.text)
Dep_t    

total_t=driver.find_elements("xpath",'//p[@class="fs-12 bold du mb-2"]')
len(total_t)

Total_t=[]
for i in total_t:
    Total_t.append(i.text)
Total_t    

total_s=driver.find_elements("xpath",'//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
len(total_s)

Total_s=[]
for i in total_s:
    Total_s.append(i.text)
Total_s    

cost=driver.find_elements("xpath",'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
len(cost)

price=[]
for i in cost:
    price.append(i.text)
price    

meal_tag=driver.find_elements("xpath",'//span[@class="text"]')
len(meal_tag)

Add_info=[]
for i in meal_tag:
    Add_info.append(i.text)
Add_info    


# Creating the Dataframe.

flights7 = pd.DataFrame({})
flights7['Airline'] = Airline
flights7['Destination'] = list1
flights7['Source'] = list2
flights7['Arrival_time'] = Arr_t
flights7['Departure_time'] =Dep_t
flights7['Total_Stops']=Total_s
flights7['Total_Duration']=duration
flights7['Additional_Info']=pd.Series(Add_info)
flights7['Price']=price
flights7
```
